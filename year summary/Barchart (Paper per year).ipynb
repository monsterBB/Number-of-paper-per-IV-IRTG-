```python
import numpy as np
import pandas as pd
from skimage import exposure,data
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("year summary.csv",encoding = 'cp1252')
print(data)
```

```python
data.DP_sum
```

```python
num_list1 = data.DP_sum
num_list2 = data.Pub_sum
num_list3 = data.PP_sum
name_list = data.year
 
plt.figure(figsize=(9,6))
n=10
index = np.arange(n)+1
bar_width = 0.4


rects1 = plt.bar(index - bar_width/2, num_list1, bar_width,
                 alpha=0.8,
                 color='skyblue',
                 label='Discussion paper') 

rects2 = plt.bar(index + bar_width/2, num_list2, bar_width,
                 alpha=0.4,
                 color='green',
                 label='Publications')
 

plt.title('Paper Per year of Principle Investigators')
plt.ylabel('Number of paper')
plt.xticks(rotation=20)
plt.xticks(index, name_list)
plt.legend()
plt.tight_layout()
plt.savefig("Paper Per year of Principle Investigators.pdf")
plt.show()

```

```python
plt.figure(figsize=(9,6))
plt.bar(index, num_list1, tick_label=name_list,alpha=0.8, fc= "skyblue")
plt.xticks(rotation=20)
plt.title('Discussion Paper Per year of Principal Investigators')
plt.savefig("Discussion Paper Per year of Principal Investigators.pdf")
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.bar(index, num_list2, tick_label=name_list,alpha=0.4, fc= 'green')
plt.xticks(rotation=20)
plt.title('Publications Per year of Principal Investigators')
plt.savefig("Publications Per year of Principal Investigators.pdf")
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.bar(index, num_list3, tick_label=name_list,alpha=0.5, fc= "red")
plt.xticks(rotation=20)
plt.title('Sum of Paper Per year of Principal Investigators')
plt.savefig("Sum of Paper Per year of Principal Investigators.pdf")
plt.show()
```
