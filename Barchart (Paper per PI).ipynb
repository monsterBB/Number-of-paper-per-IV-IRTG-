```python
import numpy as np
import pandas as pd
from skimage import exposure,data
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("result.csv",encoding = 'cp1252')
print(data)
```

```python
data.name
```

```python
num_list1 = data.DP
num_list2 = data.PP
name_list = data.abbreviate
 
plt.figure(figsize=(9,6))
n=10
index = np.arange(n)+1
bar_width = 0.4


rects1 = plt.bar(index - bar_width/2, num_list1, bar_width,
                 alpha=0.4,
                 color='blue',
                 label='Discussion paper') 

rects2 = plt.bar(index + bar_width/2, num_list2, bar_width,
                 alpha=0.8,
                 color='orange',
                 label='Publications')
 

plt.title('Paper Per Principal Investigators')
plt.ylabel('Number of paper')
plt.xticks(rotation=20)
plt.xticks(index, name_list)
plt.legend()
plt.tight_layout()
plt.savefig("Paper Per Principal Investigators.jpg")
plt.show()

```

```python
plt.figure(figsize=(9,6))
plt.bar(index, num_list1, tick_label=name_list,alpha=0.4, fc= "blue")
plt.xticks(rotation=45)
plt.title('Discussion Paper Per Principal Investigators')
plt.savefig("Discussion Paper Per Principal Investigators.jpg")
plt.show()
```

```python
plt.figure(figsize=(9,6))
plt.bar(index, num_list2, tick_label=name_list,alpha=0.8, fc= 'orange')
plt.xticks(rotation=45)
plt.title('Publications Per Principal Investigators')
plt.savefig("Publications Per Principal Investigators.jpg")
plt.show()
```
